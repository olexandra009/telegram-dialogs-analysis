5. Draw number of send and received(s&r) messages per day (x-axis - is date(from min(date) till max(date)), y-axis number of messages). There should be two lines: blue for received red for send.

```python
def get_date_only(date):
    return date.split()[0]

send_messages['date_day'] = send_messages['date'].apply(get_date_only)
received_messages['date_day'] = received_messages['date'].apply(get_date_only)
```

```python
send_group = send_messages[["id", "date_day"]].groupby("date_day", as_index= False).count()
received_group = received_messages[["id", "date_day"]].groupby("date_day", as_index= False).count()
```

```python
received_group
```

```python
send_group = send_group.sort_values("date_day")
received_group = received_group.sort_values("date_day")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(30,10))
plt.plot(received_group['date_day'], received_group['id'], label='Received', color='blue')
plt.plot(send_group['date_day'], send_group['id'], label='Sended', color='red')


plt.xlabel('Time')
plt.ylabel("Count")
plt.show()
```

7. Calculate top-10 people to whom you wrote the biggest amount of messages (name, amount of messages).

```python
df_names = df_meta[["dialog_id","name"]]
```

```python
df_names = df_names.drop_duplicates()
```

```python
def associate_dialog_with_name(dialog_id):
    return df_names[df_names["dialog_id"]==dialog_id]["name"].iloc[0]
```

```python
person_ids = df_meta[df_meta["type"]=="Private dialog"]["dialog_id"]
person_send_messages = send_messages[send_messages["dialog_id"].isin(person_ids)]
person_send_messages_group = person_send_messages[["dialog_id", "id"]].groupby("dialog_id", as_index=False).count()
person_send_messages_top_10 = person_send_messages_group.sort_values("id", ascending=False).head(10)
person_send_messages_top_10["name"] = person_send_messages_top_10["dialog_id"].apply(associate_dialog_with_name)
person_send_messages_top_10
```

4. Calculate the number of distinct user names you are connected to through any type of dialogue.

```python
def get_users_from_json(users):
    try:
        u  = users.split(",")
        u = [s for x in u for s in x.split(':')]
        id = int(u[1])
        if id==my_id:
            return None
        return id
    except:
        return None
```

```python
df_groups_and_chats_connection = df_meta[['users']] 
df_groups_and_chats_connection['dialog_id'] = df_groups_and_chats_connection['users'].apply(get_users_from_json)
df_groups_and_chats_connection = df_groups_and_chats_connection[["dialog_id"]]
```

```python
df_dialogs_connections= df_meta[df_meta["type"]=="Private dialog"][["dialog_id"]]
```

```python
concat = [df_dialogs_connections, df_groups_and_chats_connection]
```

```python
users_number = pd.concat(concat).drop_duplicates()["dialog_id"].count()
```
